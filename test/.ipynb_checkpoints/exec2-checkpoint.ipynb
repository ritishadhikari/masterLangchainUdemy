```python
import PIL
import numpy as np
import pandas as pd
from langchain_google_genai import GoogleGenerativeAI,ChatGoogleGenerativeAI
from langchain.prompts import (ChatPromptTemplate,
                               SystemMessagePromptTemplate,
                               HumanMessagePromptTemplate,
                               )
from langchain_core.messages import HumanMessage
```

```python
img=PIL.Image.open("images/3.jpg")
```

```python
img
```

```python
llmVision=ChatGoogleGenerativeAI(model="gemini-1.5-flash-001")
llmPro=ChatGoogleGenerativeAI(model="gemini-1.5-pro-latest")
```

```python
def getOCRandSummarize(filePath,llmVision,llmText):
    promptTemplateOCR=ChatPromptTemplate.from_messages(
    messages=[
        HumanMessage(
            content=[
                    {
                        'type':'text',
                        'text': """Return the texts from the images
                                """
                    },
                    {
                        'type':'image_url',
                        'image_url':filePath  # public image url, local file path, base64encoded, PIL 
                    }
                ]
            
            )
        ]
    ).format_messages()
    responseOCR=llmVision.invoke(input=promptTemplateOCR)
    print(responseOCR.content[:50])
    
    promptTemplateSummary=ChatPromptTemplate.from_messages(
    messages=[
        SystemMessagePromptTemplate.from_template(
            template="You are Text Summarization Specialist."
        ),
        HumanMessagePromptTemplate.from_template(
            template="Summarize the Text: {text} for Normal People`"
        )
    ]
    ).format_messages(text=responseOCR.content)

    responseSummary=llmText.invoke(input=promptTemplateSummary)


    return responseSummary.content
```

```python
response=getOCRandSummarize(filePath="images/3.jpg",
                            llmVision=llmVision,
                            llmText=llmPro)
```
